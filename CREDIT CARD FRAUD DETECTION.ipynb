# ML MODEL









```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a Pandas DataFrame
credit_card_data = pd.read_csv('/content/creditcard.csv')

# first 5 rows of the dataset
credit_card_data.head()
```

```python
credit_card_data.info()
```

```python
# distribution of legit & fraudulent transactions
credit_card_data['Class'].value_counts()
```

## Its a Highly unbalanced dataset

Normal Transaction ------> 0

Fraudulent transaction --> 1

```python
# separarting the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class ==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

## Comparing the values for both transactions

```python
credit_card_data.groupby('Class').mean()
```

## Under-Sampling

Building a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions



```python
legit_sample = legit.sample(n=492)  #n0.of fraudulent transactions = 492
```

```python
# concatenating two dataframes

new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

## Splitting Data into features and targets

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

### **Splitting the data into Training and Testing data:**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# **MODEL TRAINING**

## Training the Logistic Regression Model:

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

# **MODEL EVALUATION**

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on training data:',training_data_accuracy,'\n')
print('Accuracy score on testing data:', test_data_accuracy)
```
